```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('train.csv',sep = ',')
df
```

```python
df.columns
```

```python
x = df.drop('label',axis="columns")
y = df.label
```

```python
print(len(y))
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x,y)
```

```python
for i in range(2,10):
    print(f'predicted : {model.predict([x.iloc[i]])} actual : {y.iloc[i]}')
```

```python
import pickle
with open('model','wb') as f: #here wb -> write binary
    pickle.dump(model,f)
```

```python
import pickle
with open('model','rb') as f: # here rb -> read binary
		model = pickle.load(f)
model.predict([x.iloc[3]])
```
