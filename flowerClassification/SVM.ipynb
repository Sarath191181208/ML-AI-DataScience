```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data_set = load_iris()
dir(data_set)
```

```python
df = pd.DataFrame(data_set.data,columns=data_set.feature_names)
df
```

```python
df['flower'] = data_set.target
df
```

```python
data_set.target_names
```

```python
setosa_data = df[:50]
setosa_data = setosa_data.drop('flower',axis = 'columns')
setosa_data

versicolor_data = df[50:100]
versicolor_data = versicolor_data.drop('flower',axis = 'columns')



virginica_data = df[100:150]
virginica_data = virginica_data.drop('flower',axis = 'columns')
```

```python
# setosa_data,versicolor_data,virginica_data
x_label = 'sepal length (cm)'
y_label = 'petal length (cm)'

plt.xlabel(x_label)
plt.xlabel(y_label)

# support vector machine line
plt.plot([4.5,8],[4.3,5])
plt.plot([4.5,7.5],[2.3,3],color = 'red')

plt.scatter(setosa_data[x_label],setosa_data[y_label],label = 'setosa',color = 'red')
plt.scatter(versicolor_data[x_label], versicolor_data[y_label],label = 'versicolor',color = 'green')
plt.scatter(virginica_data[x_label],virginica_data[y_label],label = 'virginica',color = 'blue')

plt.legend()
```

```python
# setosa_data,versicolor_data,virginica_data

x_label = 'sepal width (cm)'
y_label = 'petal width (cm)'

plt.xlabel(x_label)
plt.xlabel(y_label)

# seperation lines for support vector machine
plt.plot([2,4],[1,2.5])
plt.plot([2,4.5],[0.5,1],color = 'red')

plt.scatter(setosa_data[x_label],setosa_data[y_label],label = 'setosa',color = 'red')
plt.scatter(versicolor_data[x_label], versicolor_data[y_label],label = 'versicolor',color = 'green')
plt.scatter(virginica_data[x_label],virginica_data[y_label],label = 'virginica',color = 'blue')

plt.legend()
```

```python
x = df.drop('flower',axis = 'columns')
y = df.flower
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.05)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
