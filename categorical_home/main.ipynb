```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv',sep=',')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged_df = pd.concat([df,dummies],axis = 'columns')
merged_df
```

dropping a town due to dummy variable trap

```python
final_df = merged_df.drop(['town','west windsor'],axis = 'columns')
final_df
```

```python
model = linear_model.LinearRegression()
x = final_df.drop('price',axis = 'columns')
y = final_df.price
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```
