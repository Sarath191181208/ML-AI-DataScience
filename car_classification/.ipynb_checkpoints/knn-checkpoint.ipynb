```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('car.csv',sep = ',')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_safety = LabelEncoder()
le_maintainance = LabelEncoder()
df.safety = le_safety.fit_transform(df.safety)
df.maintainance = le_maintainance.fit_transform(df.maintainance)
df.persons = [i if i != 'more' else 5 for i in df.persons]
df.doors = [int(i) if i != '5more' else 5 for i in df.doors]
df
```

```python
plt.scatter(df.doors,df.safety)
```
